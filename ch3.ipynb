```python
%%timeit
x = range(10000)
total = 0
for i in x:
    total += i
mean = total / len(x)
```

### Keyboard Shortcuts

I put this in a code cell at first, then used the `m` shortcut to turn it into a markdown cell.

Some other keyboard shortcuts in command mode:
- `a` inserts a code cell above current
- `b` inserts a code cell below current
- `00` restarts the kernel
- `ii` interrupts the kernel
- `x` `c` and `v` are cut, copy paste commands
- `dd` deletes current cell
- `y` change to code cell
